## Data fetching
Next we're gonna fetch several public resources (RdRp-scan, NeoRdRp, RVMT, and hopefully more in the future). If possible we'll download the MSA files and create the pHMMs from scratch, using the same version of hmmer.



```python
%%bash
mkdir DBs
cd DBs

# NeoRdRp
wget https://github.com/shoichisakaguchi/NeoRdRp/raw/main/NeoRdRp-HMM.v1.1.hmm.gz  --quiet
gzip -d NeoRdRp-HMM.v1.1.hmm.gz
sed  -e '/^NAME/s/$/_NeoRdRp/' NeoRdRp-HMM.v1.1.hmm -i # add NeoRdRp suffix to each profile

# RdRp-scan
git clone --recursive https://github.com/JustineCharon/RdRp-scan
cd RdRp-scan
mkdir RdRp-scan_HMMs
cd Profile_db_and_alignments
for alignm_file in ./ALL_NCBI_Palmdb_RdRp.*.fasta.CLUSTALO
do
ali_name=$(basename $alignm_file .fasta.CLUSTALO)
ali_name=$(echo $ali_name | cut -d "." -f2)
hmmbuild  -n "$ali_name"_RdRp-scan  -o /dev/null ../RdRp-scan_HMMs/"$ali_name".hmm $alignm_file
done

for alignm_file in ./ALL_FINAL_RdRp_cores_0.3.*.fasta.CLUSTALO
do
ali_name=$(basename $alignm_file .fasta.CLUSTALO)
ali_name=$(echo $ali_name | cut -d "." -f3)
hmmbuild  -n "$ali_name"_RdRp-scan  -o /dev/null ../RdRp-scan_HMMs/"$ali_name".hmm $alignm_file
# parallel -j"$THREADS" hmmbuild --informat a2m -n {/.} -o /dev/null ../HMMfiles/{/.}".hmm" {} ::: ./*.msa.Cons.msa.afa

done
cd ../RdRp-scan_HMMs/
cat ./*hmm  > ../RdRp-scan.hmm
cd ..
cd ..



# RVMT
### wget https://zenodo.org/record/7368133/files/RVMT_Zenodo_V4.zip?download=1
# *too big, download would take 4 hours, going to improvise:
wget https://portal.nersc.gov/dna/microbial/prokpubs/Riboviria/RiboV1.4/Alignments/zip.ali.220515.tgz  --quiet
tar -zxf zip.ali.220515.tgz
mkdir RVMT_HMMs

for ali_folder in ./ali.*
do
cd $ali_folder
for alignm_file in ./*.FASTA
do
ali_name=$(basename $alignm_file .FASTA)
hmmbuild  -n "$ali_name"_RVMT  -o /dev/null ../RVMT_HMMs/"$ali_name".hmm $alignm_file
done
cd ..
done
cd RVMT_HMMs
cat ./*hmm  > ../RVMT.hmm
cd ..


# TSA_Olendraite
## used gdrive/manual to fetch from Google Drive: https://drive.google.com/drive/folders/12rfwCDNROm-kwPRjlVOsSCzp-Gs7-wQq
## Will include the zip of ^ for prosperity...
mkdir TSA_Olendraite
cp TSA_Olendraite.zip TSA_Olendraite/TSA_Olendraite.zip
cd TSA_Olendraite
unzip TSA_Olendraite.zip
mkdir TSA_Olendraite_HMMs
for alignm_file in ./*.sto
do
ali_name=$(basename $alignm_file _consensusadded.c090.fasta.sto)
hmmbuild  -n "$ali_name"_TSA_Olendraite  -o /dev/null ./TSA_Olendraite_HMMs/"$ali_name".hmm $alignm_file
done
cd TSA_Olendraite_HMMs
cat ./*hmm  > ../TSA_Olendraite.hmm
cd ..
```

## Cleanup
Next we'll concatenate th HMM sets by the hmmer version used to create them.  
We'll also remove some intermideary files and what not...
Finally we'll write a "watermark" - text file with the date and version of the script we used (going by commit?)

```python
%%bash
cd DBs
mv ./RdRp-scan/RdRp-scan.hmm RdRp-scan.hmm
mv TSA_Olendraite/TSA_Olendraite.hmm TSA_Olendraite.hmm
rm ./RdRp-scan/ -rf

rm ali* -rf
rm TSA_Olendraite -rf
rm zip.ali.*

rm RVMT_HMMs -rf

cat RdRp-scan.hmm RVMT.hmm TSA_Olendraite.hmm > Colab_RdRp_hmmer_3.2.2.hmm
# hmmpress Colab_RdRp_hmmer_3.2.2.hmm
bzip2 -z Colab_RdRp_hmmer_3.2.2.hmm
bzip2 -z NeoRdRp-HMM.v1.1.hmm
bzip2 -z TSA_Olendraite.hmm
bzip2 -z RVMT.hmm
```
